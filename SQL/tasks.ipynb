```python
-- Combine Two Tables 
SELECT firstName, lastName, city, state
FROM Person AS p
LEFT JOIN Address AS a
ON p.personId = a.personId
```

```python
-- Second Highest Salary
SELECT IF(MIN(salary) = MAX(salary), NULL, MIN(salary)) AS SecondHighestSalary
FROM (SELECT DISTINCT salary
FROM Employee
ORDER BY salary DESC
LIMIT 2) as t;
```

```python
-- Nth Highest Salary, first try
CREATE FUNCTION getNthHighestSalary(N INT) RETURNS INT
BEGIN
  RETURN (
            SELECT IF(N > (SELECT COUNT(salary) FROM (SELECT DISTINCT salary
            FROM Employee) as t) OR N < 0, NULL, MIN(salary)) AS getNthHighestSalary
            FROM (SELECT DISTINCT salary
            FROM Employee
            ORDER BY salary DESC
            LIMIT N) as t
  );
END
```

```python
-- Nth Highest Salary, final try
-- i could make the if condition simpler a looot
CREATE FUNCTION getNthHighestSalary(N INT) RETURNS INT
BEGIN
  RETURN (
      SELECT IF(COUNT(*) < N  OR N < 0, NULL, MIN(salary))
      FROM (
          SELECT DISTINCT salary
          FROM Employee
          ORDER BY salary DESC
          LIMIT N
      ) AS t
  );
END
```

```python
-- Rank Scores
SELECT score,
       DENSE_RANK() OVER (ORDER BY score DESC) AS "rank"
FROM Scores;
```

```python
-- Consecutive Numbers
SELECT DISTINCT A.Num AS ConsecutiveNums
FROM Logs AS A
INNER JOIN Logs AS B ON A.Id + 1 = B.Id
INNER JOIN Logs AS C ON B.Id + 1 = C.Id
WHERE A.Num = B.Num AND B.Num = C.Num
```

```python
-- Employees Earning More Than Their Managers
SELECT A.Name AS Employee
FROM Employee AS A
INNER JOIN Employee AS B
ON A.managerId = B.id
WHERE A.salary > B.salary
```

```python
-- Duplicate Emails
SELECT email
FROM Person
GROUP BY email
HAVING COUNT(*) > 1;
```

```python
-- Customers Who Never Order
SELECT c.name AS Customers
FROM Customers AS c
LEFT JOIN Orders AS o
ON c.id = o.customerId
WHERE o.customerId is null
```

```python
-- Department Highest Salary
SELECT d.name AS Department,
       e.name AS Employee,
       e.salary AS Salary
FROM Employee AS e
JOIN Department AS d
  ON e.departmentId = d.id
WHERE e.salary = (
    SELECT MAX(salary)
    FROM Employee
    WHERE departmentId = e.departmentId
);
```

```python
-- Exchange Seats
SELECT ( CASE
            WHEN id % 2 != 0 AND id != counts THEN id + 1
            WHEN id % 2 != 0 AND id = counts THEN id
            ELSE id - 1
        END) AS id, student
FROM Seat, (SELECT count(*) AS counts FROM seat) AS t
ORDER BY id ASC
```
