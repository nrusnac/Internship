```python
# Two Sum
class Solution:
    def twoSum(self, nums, target):
        seen = {}

        for i in range(len(nums)):
            complement = target - nums[i]

            if complement in seen:
                return [seen[complement], i]

            seen[nums[i]] = i
```

```python
# Contains Duplicate
from typing import List

class Solution:
    def hasDuplicate(self, nums: List[int]) -> bool:
        nums.sort()
        for i in range(len(nums)-1):
            if nums[i] == nums[i+1]:
                return True
        return False
```

```python
# Best Time to Buy and Sell Stock
class Solution:
    def maxProfit(self, prices: List[int]) -> int:
        min_price = float('inf')
        max_profit = 0

        for price in prices:
            if price < min_price:
                min_price = price
            elif price - min_price > max_profit:
                max_profit = price - min_price
        return max_profit
```

```python
# Valid Anagram
class Solution:
    def isAnagram(self, s: str, t: str) -> bool:
        set1 = {}
        set2 = {}

        for word in s:
            if word not in set1.keys():
                set1[word] = 1
            else:
                set1[word] += 1

        for word in t:
            if word not in set2.keys():
                set2[word] = 1
            else:
                set2[word] += 1

        if set1 == set2:
            return True

        return False
```

```python
# Valid Parentheses
class Solution:
    def isValid(self, s: str) -> bool:
        if len(s) % 2:
            return False
            
        stack = []

        for par in s:
            if par == '[' or par == '{' or par == '(':
                stack.append(par)
                continue

            if par == ']': 
                if len(stack) == 0:
                    return False
                if stack.pop() != '[':
                    return False

            if par == '}':
                if len(stack) == 0:
                    return False
                if stack.pop() != '{':
                    return False

            if par == ')': 
                if len(stack) == 0:
                    return False
                if stack.pop() != '(':
                    return False
            
        if len(stack) != 0:
            return False
        return True
```

```python
# Product of Array Except Self
class Solution:
    def productExceptSelf(self, nums: List[int]) -> List[int]:
        if all(nums):
            prod = 1
            for num in nums:
                prod *= num
            return [prod // num for num in nums]
        elif nums.count(0) == 1:
            prod = 1
            for num in nums:
                if num != 0:
                    prod *= num
            return [prod if num == 0 else 0 for num in nums]
        else:
            return [0] * len(nums)
        
```

```python
# Maximum Subarray first try
class Solution:
    def maxSubArray(self, nums: List[int]) -> int:
        maximum = float('-inf')
        if not nums:
            return maximum

        while nums:
            new_max = sum(nums)

            if new_max > maximum:
                maximum = new_max
            
            if nums[0] > nums[-1]:
                nums.pop()
            elif nums[0] < nums[-1]:
                nums.pop(0)
            else:
                left_max = self.maxSubArray(nums[1:])
                right_max = self.maxSubArray(nums[:-1])
                maximum = max(maximum, left_max, right_max)
                break

        return maximum
```

```python
# Maximum Subarray final version
class Solution:
    def maxSubArray(self, nums: List[int]) -> int:
        best = current = nums[0]
        for x in nums[1:]:
            current = max(x, current + x)
            best = max(best, current)
        return best
```

```python
# 3Sum
class Solution:
    def threeSum(self, nums: List[int]) -> List[List[int]]:
        res = []
        nums.sort()

        for i, a in enumerate(nums):
            if a > 0:
                break

            if i > 0 and a == nums[i - 1]:
                continue

            l, r = i + 1, len(nums) - 1
            while l < r:
                threeSum = a + nums[l] + nums[r]
                if threeSum > 0:
                    r -= 1
                elif threeSum < 0:
                    l += 1
                else:
                    res.append([a, nums[l], nums[r]])
                    l += 1
                    r -= 1
                    while nums[l] == nums[l - 1] and l < r:
                        l += 1

        return res
```

```python
# Merge intervals
class Solution(object):
    def merge(self, intervals):
        if not intervals:
            return []

        intervals.sort(key=lambda x: x[0])
        merged = [intervals[0][:]] 

        for start, end in intervals[1:]:
            if start <= merged[-1][1]:
                merged[-1][1] = max(merged[-1][1], end)
            else:
                merged.append([start, end])

        return merged
```

```python
# Group Anagrams first try
class Solution:
    def isAnagram(self, s: str, t: str) -> bool:
        set1 = {}
        set2 = {}

        for word in s:
            if word not in set1.keys():
                set1[word] = 1
            else:
                set1[word] += 1

        for word in t:
            if word not in set2.keys():
                set2[word] = 1
            else:
                set2[word] += 1

        if set1 == set2:
            return True
        return False
    
    def groupAnagrams(self,strs_: List[str]) -> List[List[str]]:
        strs = strs_.copy()
        angr = [[strs.pop(0)]]
        for str in strs:
            added = False
            for an in angr:
                if self.isAnagram(an[0], str):
                    an.append(str)
                    added = True
                    break
            if not added:
                angr.append([str])
        return angr
```

```python
# Group Anagrams final try
class Solution:
    def computeAnagram(self, s: str) -> str:
        set = {}
        for c in s:
            if c in set.keys():
                set[c] += 1
            else:
                set[c] = 1

        return frozenset(set.items())
    
    def groupAnagrams(self, strs_):
        strs = strs_.copy()
        groups = {}
        for str in strs:
            angr = self.computeAnagram(str)
            if angr in groups.keys():
                groups[angr].append(str)
            else:
                groups[angr] = [str]
        return [group for group in groups.values()]
```

```python
# Reverse Linked list
from typing import Optional

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution:
    def reverseList(self, head: Optional[ListNode]) -> Optional[ListNode]:
        prev = None
        curr = head
        while curr:
            nxt = curr.next
            curr.next = prev
            prev = curr
            curr = nxt
 
        return prev
```

```python
# Linked List Cycle Detection first try
class Solution:
    def hasCycle(self, head: Optional[ListNode]) -> bool:
        seen = set()
        curr = head

        while curr:
            if curr in seen:
                return True
            seen.add(curr)
            curr = curr.next

        return False
```

```python
# Linked List Cycle Detection final try
class Solution:
    def hasCycle(self, head: Optional[ListNode]) -> bool:
        p1 = head
        p2 = head

        while p2:
            p2 = p2.next
            if not p2:
                return False
            p2 = p2.next
            p1 = p1.next

            if p1 == p2:
                return True
            
        return False
```

```python
# Container With Most Water
class Solution:
    def maxArea(self, heights: List[int]) -> int:
        max_vol = 0
        s, e = 0, len(heights) - 1

        while s < e:
            p1 = heights[s]
            p2 = heights[e]
            new_vol = min(p1, p2) * (e - s)
            if new_vol > max_vol:
                max_vol = new_vol
            
            if p1 <= p2:
                s += 1
            else:
                e -= 1

        return max_vol
```

```python
# Find Minimum in Rotated Sorted Array
class Solution:
    def findMin(self, nums: List[int]) -> int:
        s, e = 0, len(nums) - 1

        while s < e:
            m = s + (e - s) // 2
            if nums[m] > nums[e]:
                s = m + 1
            else:
                e = m

        return nums[s]
```

```python
# Longest Repeating Character Replacement
class Solution:
    def characterReplacement(self, s: str, k: int) -> int:
        count = {}
        res = 0

        l = 0
        maxf = 0
        for r in range(len(s)):
            count[s[r]] = 1 + count.get(s[r], 0)
            maxf = max(maxf, count[s[r]])

            while (r - l + 1) - maxf > k:
                count[s[l]] -= 1
                l += 1
            res = max(res, r - l + 1)

        return res
```

```python
# Longest Substring Without Repeating Characters
class Solution:
    def lengthOfLongestSubstring(self, s: str) -> int:
        seen = []
        max_len = 0
        for word in s:
            if word not in seen:
                seen.append(word)
            else:
                if len(seen) > max_len:
                    max_len = len(seen)
                
                seen = seen[seen.index(word) + 1:]
                seen.append(word)
        
        if len(seen) > max_len:
            max_len = len(seen)
        return max_len
```

```python
# Number of Islands
class DSU:
    def __init__(self, n):
        self.Parent = list(range(n + 1))
        self.Size = [1] * (n + 1)

    def find(self, node):
        if self.Parent[node] != node:
            self.Parent[node] = self.find(self.Parent[node])
        return self.Parent[node]

    def union(self, u, v):
        pu = self.find(u)
        pv = self.find(v)
        if pu == pv:
            return False
        if self.Size[pu] >= self.Size[pv]:
            self.Size[pu] += self.Size[pv]
            self.Parent[pv] = pu
        else:
            self.Size[pv] += self.Size[pu]
            self.Parent[pu] = pv
        return True

class Solution:
    def numIslands(self, grid: List[List[str]]) -> int:
        ROWS, COLS = len(grid), len(grid[0])
        dsu = DSU(ROWS * COLS)

        def index(r, c):
            return r * COLS + c

        directions = [(1, 0), (-1, 0), (0, 1), (0, -1)]
        islands = 0

        for r in range(ROWS):
            for c in range(COLS):
                if grid[r][c] == '1':
                    islands += 1
                    for dr, dc in directions:
                        nr, nc = r + dr, c + dc
                        if (nr < 0 or nc < 0 or nr >= ROWS or
                            nc >= COLS or grid[nr][nc] == "0"):
                            continue

                        if dsu.union(index(r, c), index(nr, nc)):
                            islands -= 1

        return islands
```

```python
# Remove Node From End of Linked List
class Solution:
    def removeNthFromEnd(self, head: Optional[ListNode], n: int) -> Optional[ListNode]:
        p1 = head
        p2 = head

        for _ in range(n):
            p2 = p2.next

        # deleting the head node
        if p2 is None:
            return head.next

        while p2.next is not None:
            p1 = p1.next
            p2 = p2.next

        p1.next = p1.next.next
        return head
```

```python
# Palindromic Substrings
class Solution:
    def countSubstrings(self, s: str) -> int:
        n = len(s)
        res = n

        for i in range(n):
            
            l, r = i - 1, i + 1
            while l >= 0 and r < n:
                if s[l] == s[r]:
                    l -= 1
                    r += 1
                    res += 1
                else:
                    break
            
            l, r = i, i + 1
            while l >= 0 and r < n:
                if s[l] == s[r]:
                    l -= 1
                    r += 1
                    res += 1
                else:
                    break
        return res
```

```python
# Symmetric Tree
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

class Solution:
    def isSymmetric(self, root: Optional[TreeNode]) -> bool:
        if not root:
            return True
        
        queue = [(root.left, root.right)]

        while queue:
            l, r = queue.pop(0)

            if l is None and r is None:
                continue

            if l is None or r is None:
                return False
            if l.val != r.val:
                return False
            
            queue.append((l.left, r.right))
            queue.append((l.right, r.left))

        return True

```

```python
# Path Sum
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

class Solution:
    def hasPathSum(self, root: Optional[TreeNode], targetSum: int) -> bool:
        if not root:
            return False
        
        queue = [(root, targetSum - root.val)]

        while queue:
            node, curr = queue.pop(0)

            if (node.left is None and node.right is None) and (curr == 0):
                return True

            if node.left:
                queue.append((node.left, curr - node.left.val))
            if node.right:
                queue.append((node.right, curr - node.right.val))

        return False
```
